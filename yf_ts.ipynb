```python
# Now hourly data
day_df = pd.DataFrame()
hour_df = pd.DataFrame()


for symbol in big_tech.keys():
    print(f"Fetching {symbol}...")
    ticker = yf.Ticker(symbol)
    hist = ticker.history(period='1d',  interval="60m")
    
    # Create DataFrame with date as a column
    tmp_df = pd.DataFrame({
        'Stock': symbol,
        'Date': hist.index,  # ← This was missing!
        'Open': hist['Open'],
        'High': hist['High'],    # ← Add this for better analysis
        'Low': hist['Low'],      # ← Add this too
        'Close': hist['Close'],
        'Volume': hist['Volume']
    })
    
    day_df = pd.concat([day_df, tmp_df], ignore_index=True)
print('Done')


for symbol in big_tech.keys():
    print(f"Fetching {symbol}...")
    ticker = yf.Ticker(symbol)
    hist = ticker.history(period='1d',  interval="5m")
    
    # Create DataFrame with date as a column
    tmp_df = pd.DataFrame({
        'Stock': symbol,
        'Date': hist.index,  # ← This was missing!
        'Open': hist['Open'],
        'High': hist['High'],    # ← Add this for better analysis
        'Low': hist['Low'],      # ← Add this too
        'Close': hist['Close'],
        'Volume': hist['Volume']
    })
    
    hour_df = pd.concat([hour_df, tmp_df], ignore_index=True)
print('Done')

now = dt.datetime.now(dt.timezone(dt.timedelta(hours=-4)))  # Adjust for your timezone if you want
current_hour_start = now.replace(minute=0, second=0, microsecond=0)
hour_df = hour_df[(hour_df['Date'] >= current_hour_start) & (hour_df['Date'] < current_hour_start + dt.timedelta(hours=1))]


def calculate_metrics(df):
    """Clean function to calculate all metrics"""
    df = df.sort_values(['Stock', 'Date']).copy()
    
    # Basic calculations
    df['Daily_Change'] = ((df['Close'] - df['Open']) / df['Open']) * 100
    df['Daily_Range'] = ((df['High'] - df['Low']) / df['Open']) * 100
    
    # Returns and volatility
    df['Daily_Return'] = df.groupby('Stock')['Close'].pct_change() * 100
    df['Volatility_20d'] = df.groupby('Stock')['Daily_Return'].transform(
        lambda x: x.rolling(20, min_periods=5).std()
    )
    
    return df

day_df = calculate_metrics(day_df)
hour_df = calculate_metrics(hour_df)

```
